# 반복문 안에서의 return

```python
def return_in_for():
    for n in range(10):
        return n
```

`return`의 의미:
*   값을 함수를 호출한 곳으로 반환.
*   함수를 종료.

```python
result = return_in_for()
print(result)
```

# 반복문 안에서의 yield

```python
def yield_in_for():
    for n in range(10):
        yield n
```

```python
result = yield_in_for()
print(result)
```

*   `yield`를 사용하는 함수의 리턴값은 `generator` 객체 타입.
*   `generator` 객체는 for-in 구문에서 사용하는 iterable(반복이 가능한) 객체.

```python
for x in result:
    print(x)
```

```python
for x in result:
    print(x)
```

* `generator` 객체는 for-in 구문을 사용해서 yield되는 값들을 순서대로 뽑아낼 수 있음.
* yield되는 값들은 한번 뽑아내면 메모리에서 삭제됨.
* 첫번째 for-in 구문에서는 yield되는 값들을 순서대로 출력.
* 두번째 for-in 구문에서는 yield된 값들이 한번씩 사용이 됐기 때문에, 출력문이 없음!

```python
for x in yield_in_for():
    print(x)
```

```python
def my_generator():
    values = ['Google', 'Meta', 'OpenAI', 'Anthropic']
    for v in values:
        yield v
```

```python
for x in my_generator():
    print(x)
```
