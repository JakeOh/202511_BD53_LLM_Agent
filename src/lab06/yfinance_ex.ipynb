```python
samsung = yf.Ticker('005930.KS')
```

```python
print(type(samsung))  #> Ticker 클래스 객체
```

```python
samsung.info  #> Ticker 객체의 info 속성: dict
```

```python
samsung.info.keys()
```

```python
# Ticker.history(period): 주가 데이터를 반환
# period 파라미터: 주가를 검색할 기간. (예) '5d': 5일간, '1mo': 1개월간, '1y': 1년간
history_5d = samsung.history('5d')
```

```python
type(history_5d)  #> DataFrame
```

```python
history_5d
```

```python
samsung.history(period='1mo')
```
